```python
import sys
sys.path.insert(0, '../..')
from dependencies import *
from constants import *
from paths import *
import helper_functions
helper_functions.set_plot_style()
```

```python
SUBJECTS = helper_functions.alice_get_subjects()
```

```python
# -----------------------------
# Load TRFs (h NDVars)
# -----------------------------
encoder_log_trf = []
encoder_onset_trf = []
decoder_log_trf = []
decoder_onset_trf = []

universal_encoder_log_trf = eelbrain.load.unpickle(ALICE_TRF_DIR / f'universal-trf-envelope_log.pickle')
universal_encoder_onset_trf = eelbrain.load.unpickle(ALICE_TRF_DIR / f'universal-trf-envelope_onset.pickle')

for subject in SUBJECTS:
    encoder_log_trf.append(eelbrain.load.unpickle(ALICE_TRF_DIR / subject / f'{subject} envelope_log.pickle'))
    encoder_onset_trf.append(eelbrain.load.unpickle(ALICE_TRF_DIR / subject / f'{subject} envelope_onset.pickle'))
    decoder_log_trf.append(eelbrain.load.unpickle(ALICE_TRF_DIR / subject / f'{subject} decoder-envelope_log.pickle'))
    decoder_onset_trf.append(eelbrain.load.unpickle(ALICE_TRF_DIR / subject / f'{subject} decoder-envelope_onset.pickle'))
# --- Parameters ---
latency_times = [0.04, 0.14, 0.24]
times_log = [0.04, 0.14, 0.24]
times_onset = [0.05, 0.2, 0.32]
time_labels = ['%d ms' % (t*1000) for t in times_log]
```

```python
# ------------------------------------------------
# Encoder TRFs and subject lists
# ------------------------------------------------
encoder_trfs = {
    'log': universal_encoder_log_trf,
    'onset': universal_encoder_onset_trf
}

all_subjects_dict = {
    'log': encoder_log_trf,
    'onset': encoder_onset_trf
}

times_dict = {
    'log': times_log,
    'onset': times_onset
}

# ------------------------------------------------
# Compute variability across subjects
# ------------------------------------------------
variability_maps = {}
mean_trfs = {}
for trf_type in ['log', 'onset']:
    subjects = all_subjects_dict[trf_type]
    subject_trfs = []
    subject_var_expl = []

    for sub in subjects:
        if isinstance(sub, tuple):
            trf = sub[0]  # unwrap tuple -> BoostingResult
            subject_trfs.append(trf.h[0] if isinstance(trf.h, tuple) else trf.h)
            subject_var_expl.append(trf.proportion_explained)
        elif isinstance(sub, eelbrain.BoostingResult):
            subject_trfs.append(sub.h[0] if isinstance(sub.h, tuple) else sub.h)
            subject_var_expl.append(sub.proportion_explained)
        else:
            raise ValueError(f"Unexpected subject type: {type(sub)}")

    
    trfs = eelbrain.combine(subject_trfs)
    vars_explained = eelbrain.combine(subject_var_expl)


    mean_trfs[trf_type] = trfs.mean('case')
    variability_maps[trf_type] = vars_explained.mean("case")

    

# ------------------------------------------------
# Figure setup
# ------------------------------------------------
fig = plt.figure(figsize=(15, 9))

# Panel labels
fig.text(0.05, 0.96, 'A) Predictive power', fontsize=16, fontweight='bold')
fig.text(0.50, 0.96, 'B) Encoder TRF', fontsize=16, fontweight='bold')

# Layout parameters
left_col = 0.05
right_col = 0.30
row_height = 0.40
gap_row = 0.04
width_var = 0.15
height_topo = 0.15
width_butterfly = 0.60
width_topo = 0.10
gap_topo = 0.005  # decreased space between topomaps

# ------------------------------------------------
# Loop over models
# ------------------------------------------------
for row_idx, trf_type in enumerate(['log', 'onset']):
    latency_times = times_dict[trf_type]

    trf = mean_trfs[trf_type]
    trf_vmax = trf.max()
    var_ndvar = variability_maps[trf_type]

    # Vertical position for this row
    bottom_row = 0.52 - row_idx*(row_height + gap_row)

    # ---------------------
    # Row-level title (centered above both left and right)
    # ---------------------
    left_edge = left_col
    right_edge = right_col + width_butterfly
    type_model = f'Acoustic Envelope' if trf_type == 'log' else f'Acoustic Onset'
    fig.text((left_edge + right_edge)/2 -0.15, bottom_row + row_height - 0.05,
             f'{type_model} TRF', fontsize=16, fontweight='bold',
             ha='center', va='bottom')

    # ---------------------
    # Variability topomap (left)
    # ---------------------
    ax_var = fig.add_axes([left_col, bottom_row + (row_height - height_topo)/2, width_var, height_topo*1.2])
    eelbrain.plot.Topomap(
        var_ndvar * 100, # Get in percentage
        axes=ax_var,
        axtitle='Variability explained across subjects',
        cmap='RdBu_r',
        clip='circle',
        vmax=None,
    )
    im_var = ax_var.get_images()[0]
    cbar_ax = fig.add_axes([
        ax_var.get_position().x0,
        ax_var.get_position().y0 - 0.025,
        ax_var.get_position().width,
        0.02
    ])
    fig.colorbar(im_var, cax=cbar_ax, orientation='horizontal', label="% variability explained")

    # ---------------------
    # Butterfly TRF (right)
    # ---------------------
    ax_butterfly = fig.add_axes([right_col, bottom_row, width_butterfly, row_height*0.7])
    eelbrain.plot.Butterfly(
        trf,
        axes=ax_butterfly,
        linewidth=0.8,
        color='tab:blue' if trf_type == 'log' else 'tab:orange',
        ylabel='TRF weights [a.u.]',
        yticklabels='none',
        frame='t'
    )
    ax_butterfly.set_xlim(trf.time[0], trf.time[-1])
    y_lim = [trf.x.min(), trf.x.max()]
    print(y_lim)
    ax_butterfly.set_ylim(y_lim[0] * 2, y_lim[1] * 2)

    # Add vertical red lines for the latency times
    for t in latency_times:
        ax_butterfly.axvline(t, color='r', linestyle='--', linewidth=0.8)

    # ---------------------
    # Latency topomaps (top-right, above butterfly)
    # ---------------------
    n_topos = len(latency_times)
    butterfly_pos = ax_butterfly.get_position()
    bottom_topos = butterfly_pos.y0 + 0.3*butterfly_pos.height + 0.01
    left_start = butterfly_pos.x0 + butterfly_pos.width - (n_topos*width_topo + (n_topos-1)*gap_topo)

    topomap_axes = []
    for i, t in enumerate(latency_times):
        ax_topo = fig.add_axes([
            left_start + i*(width_topo + gap_topo) - i*0.015,
            bottom_topos,
            width_topo,
            height_topo
        ])
        topo_trf = trf.sub(time=t)
        plot_topo = eelbrain.plot.Topomap(
            topo_trf,
            axes=ax_topo,
            axtitle=f"{int(t*1000)} ms",
            clip='circle',
            vmax=None,
        )
        topomap_axes.append(ax_topo)
        


    # ---------------------
    # Add vertical colorbar to the right of the topomaps
    # ---------------------
    im_topo = topomap_axes[-1].get_images()[0]
    pos = topomap_axes[-1].get_position()
    cbar_ax = fig.add_axes([
        pos.x0 + pos.width + 0.015,  # just to the right
        bottom_topos,
        0.01,                        # vertical bar width
        height_topo,                    # same height as topomaps
    ])
    cbar = fig.colorbar(im_topo, cax=cbar_ax, orientation='vertical')
    vmin, vmax = im_topo.get_clim()
    cbar.set_ticks([vmin, 0, vmax])
    cbar.set_ticklabels(['-', '0', '+'])
    
plt.savefig(ALICE_FIGURES_DIR / 'alice_forward_trfs.png', dpi=300)
plt.show()
```
