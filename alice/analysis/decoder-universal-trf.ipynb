```python
import sys
sys.path.insert(0, '../..')
from dependencies import *
from constants import *
from paths import *
import helper_functions
```

```python
SUBJECTS = helper_functions.alice_get_subjects()
models = helper_functions.alice_get_models(exclude=['envelope_log_onset'])
```

```python
# Loop over all models
for model in models:
    print(f'\nProcessing universal TRF for model: {model}')
    trf_list = []

    # Collect TRFs for all subjects for this model
    for subject in SUBJECTS:
        trf = eelbrain.load.unpickle(ALICE_TRF_DIR / subject / f'{subject} decoder-{model}.pickle').h_scaled
        trf_list.append(trf)

    # Combine and average across subjects
    universal_trf = eelbrain.combine(trf_list).mean('case')

    # Save universal TRF for this model
    eelbrain.save.pickle(universal_trf, ALICE_TRF_DIR / f'decoder-universal-trf-{model}.pickle')

    print(f'Saved universal TRF for model {model}')

    print(trf_list[0])
```
