```python
# DATA PROCESSING OF ENVELOPES

STIMULI = [str(i) for i in range(1, 13)]


# Resample spectrograms to 100 hz, offset and filter (data processing)
# Log
envelopes_log = [eelbrain.load.unpickle(ALICE_PREDICTOR_DIR / f'{stimulus}~envelope-log.pickle') for stimulus in STIMULI]
envelopes_log = [x.bin(0.01, dim='time', label='start') for x in envelopes_log]
envelopes_log = [eelbrain.pad(x, tstart=-0.100, tstop=x.time.tstop + 1, name='envelope_log') for x in envelopes_log]
envelopes_log = [eelbrain.filter_data(x, BANDPASS_FILTER_LOW, BANDPASS_FILTER_HIGH) for x in envelopes_log]
# Onset
envelopes_onset = [eelbrain.load.unpickle(ALICE_PREDICTOR_DIR / f'{stimulus}~envelope-onset.pickle') for stimulus in STIMULI]
envelopes_onset = [x.bin(0.01, dim='time', label='start') for x in envelopes_onset]
envelopes_onset = [eelbrain.pad(x, tstart=-0.100, tstop=x.time.tstop + 1, name='envelope_onset') for x in envelopes_onset]
envelopes_onset = [eelbrain.filter_data(x, BANDPASS_FILTER_LOW, BANDPASS_FILTER_HIGH) for x in envelopes_onset]
```

```python
# ————————————————————————————————————————————————————————————————————————————————————————————————
# Save pickles

eelbrain.save.pickle(envelopes_log, 
                        ALICE_PROCESSED_PREDICTOR_DIR / f'~processed_envelopes-log.pickle') # <-- This is referred to as gammatone-1 in document
eelbrain.save.pickle(envelopes_onset, 
                        ALICE_PROCESSED_PREDICTOR_DIR / f'~processed_envelopes-onset.pickle') # <-- This is referred to as gammatone-on-1 in document
print(envelopes_log)
```
