```python
# ─────────────────────────────────────────────
# Load TRFs for butterfly plots
# ─────────────────────────────────────────────

def load_general_trf(predictor, attention, generalised):
    name = helper_functions.get_trf_model_name(
        DATASET_TYPE.FUGLSANG, predictor, attention, MODEL_TYPE.BACKWARD, generalised=generalised
    )
    return eelbrain.load.unpickle(FUGLSANG_TRF_GENERAL_DIR / f'full_{name}.pickle')

trf_avg_env    = load_general_trf(PREDICTOR_TYPE.ENVELOPE,       ATTENTION_TYPE.ATTENDED, GENERALISATION_TYPE.AVERAGE)
trf_pool_env   = load_general_trf(PREDICTOR_TYPE.ENVELOPE,       ATTENTION_TYPE.ATTENDED, GENERALISATION_TYPE.POOLED)
trf_avg_onset  = load_general_trf(PREDICTOR_TYPE.ENVELOPE_ONSET, ATTENTION_TYPE.ATTENDED, GENERALISATION_TYPE.AVERAGE)
trf_pool_onset = load_general_trf(PREDICTOR_TYPE.ENVELOPE_ONSET, ATTENTION_TYPE.ATTENDED, GENERALISATION_TYPE.POOLED)

# ─────────────────────────────────────────────
# Figure 1 — Butterfly + Topomap
# ─────────────────────────────────────────────

latencies_envelope  = [-0.180, -0.140, -0.075]
latencies_onset     = [-0.250, -0.140, -0.080]

butterfly_rows = [
    (trf_avg_env,    'Averaged — Attended Envelope',  'tab:blue',   latencies_envelope),
    (trf_pool_env,   'Pooled — Attended Envelope',    'tab:blue',   latencies_envelope),
    (trf_avg_onset,  'Averaged — Attended Onset',     'tab:orange', latencies_onset),
    (trf_pool_onset, 'Pooled — Attended Onset',       'tab:orange', latencies_onset),
]

max_topos  = len(latencies_envelope)
n_rows     = len(butterfly_rows)

all_trfs_prepared = [trf.h_scaled for trf, _, _, _ in butterfly_rows]

# Per-predictor scales
def group_scale(trfs):
    abs_max = max(max(abs(t.x.min()), abs(t.x.max())) for t in trfs)
    return abs_max * 1.5, abs_max

env_ylim,   env_vmax   = group_scale(all_trfs_prepared[:2])
onset_ylim, onset_vmax = group_scale(all_trfs_prepared[2:])

scales = [
    (env_ylim,   env_vmax),
    (env_ylim,   env_vmax),
    (onset_ylim, onset_vmax),
    (onset_ylim, onset_vmax),
]

fig1 = plt.figure(figsize=(16, 3 * n_rows))
gs1  = fig1.add_gridspec(
    n_rows, 1 + max_topos,
    width_ratios=[4] + [1] * max_topos,
    hspace=0.6, wspace=0.3
)

for i, (trf, title, color, latencies) in enumerate(butterfly_rows):
    trf_plot   = all_trfs_prepared[i]
    ylim, vmax = scales[i]

    ax_trf = fig1.add_subplot(gs1[i, 0])
    eelbrain.plot.Butterfly(
        trf_plot,
        axes=ax_trf,
        color=color,
        linewidth=0.8,
        frame='t',
        ylabel='TRF weights (a.u.)',
        xlabel='Time lag (s)' if i == n_rows - 1 else '',
        xticklabels='none' if i < n_rows - 1 else True,
    )

    ax_trf.set_ylim(-ylim, ylim)
    ax_trf.axhline(0, linestyle='--', color='black', linewidth=0.8)
    ax_trf.axvline(0, linestyle='--', color='black', linewidth=0.8)
    ax_trf.grid(True, linestyle='--', alpha=0.4)
    ax_trf.set_title(title)
    ax_trf.set_yticks([])

    for t in latencies:
        ax_trf.axvline(t, color='red', linestyle='--', linewidth=1)

    topo_axes = [fig1.add_subplot(gs1[i, 1 + j]) for j in range(max_topos)]
    t_labels  = [f'{int(t * 1000)} ms' for t in latencies]

    plot_topo = eelbrain.plot.Topomap(
        [trf_plot.sub(time=t) for t in latencies],
        axes=topo_axes,
        axtitle=t_labels,
        columns=max_topos,
        vmax=vmax,
        clip='circle',
    )
    plot_topo.plot_colorbar(
        right_of=topo_axes[-1], label='', label_rotation=90,
        ticks={vmax: '+', -vmax: '-', 0: '0'}
    )

fig1.suptitle('Universal Backward TRFs — Attended (Averaged vs Pooled)', y=1.01)
fig1.savefig(BEYOND_FIGURES_DIR / 'beyond_pooled_butterfly_topomap.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# ─────────────────────────────────────────────
# Figure 2 — Scatter panels
# ─────────────────────────────────────────────

def plot_scatter_panel(ax, subjects, 
                       r_att_avg, r_ign_avg, acc_avg,
                       r_att_pool, r_ign_pool, acc_pool,
                       color, title):

    r_att_avg  = np.array(list(r_att_avg.values()))
    r_ign_avg  = np.array(list(r_ign_avg.values()))
    r_att_pool = np.array(list(r_att_pool.values()))
    r_ign_pool = np.array(list(r_ign_pool.values()))

    for r_att, r_ign, acc, dot_color, marker, label in [
        (r_att_avg,  r_ign_avg,  acc_avg, "black",  'o', f'Averaged ({acc_avg:.0%})'),
        (r_att_pool, r_ign_pool, acc_pool, color, 'o', f'Pooled ({acc_pool:.0%})'),
    ]:
        correct   = r_att > r_ign
        incorrect = ~correct

        ax.scatter(r_att[correct], r_ign[correct],
                   color=dot_color, edgecolors='white', linewidths=0.4,
                   s=65, zorder=3, marker=marker, label=f'{label} ✓')
        ax.scatter(r_att[incorrect], r_ign[incorrect],
                   color=dot_color, edgecolors='red', linewidths=0.7,
                   s=65, zorder=3, marker="X",
                   label=f'{label} ✗')

        for s, x, y in zip(subjects, r_att, r_ign):
            ax.annotate(s, (x, y), textcoords='offset points', xytext=(5, 3),
                        fontsize=8, color=dot_color, alpha=0.85)

    all_r = np.concatenate([r_att_avg, r_ign_avg, r_att_pool, r_ign_pool])
    lim   = [-0.02, max(all_r) * 1.2]
    ax.plot(lim, lim, 'k--', linewidth=0.9, alpha=0.5, zorder=2)
    ax.set_xlim(lim)
    ax.set_ylim(lim)
    ax.set_aspect('equal')
    ax.set_xlabel(r'$r_{att}$')
    ax.set_ylabel(r'$r_{ign}$')
    ax.set_title(title, fontsize=12)
    ax.legend(loc='upper left', framealpha=0.9,)
    ax.spines[['top', 'right']].set_visible(False)
    ax.tick_params(labelsize=8)


fig2, (ax_env, ax_onset) = plt.subplots(1, 2, figsize=(14, 6))


plot_scatter_panel(
    ax_env, SUBJECTS,
    r_atts_avg_env, r_igns_avg_env, acc_avg_env,
    r_atts_pool_env, r_igns_pool_env, acc_pool_env,
    color='tab:blue',
    title='Envelope\nAveraged vs Pooled',
)

plot_scatter_panel(
    ax_onset, SUBJECTS,
    r_atts_avg_onset, r_igns_avg_onset, acc_avg_onset,
    r_atts_pool_onset, r_igns_pool_onset, acc_pool_onset,
    color='tab:orange',
    title='Onset\nAveraged vs Pooled',
)



fig2.suptitle('AAD Classification: Averaged vs Pooled LOOCV')
fig2.tight_layout()
fig2.savefig(BEYOND_FIGURES_DIR / 'beyond_aad_pooled_scatter.png', dpi=150, bbox_inches='tight')
plt.show()

```
