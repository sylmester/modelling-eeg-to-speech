# Estimate averaged universal TRFs

## Full 18-subject averaged TRFs

```python
# TRF types for analysis
trf_types = [
    (PREDICTOR_TYPE.ENVELOPE,       ATTENTION_TYPE.ATTENDED, MODEL_TYPE.BACKWARD),
    (PREDICTOR_TYPE.ENVELOPE,       ATTENTION_TYPE.IGNORED,  MODEL_TYPE.BACKWARD),
    (PREDICTOR_TYPE.ENVELOPE_ONSET, ATTENTION_TYPE.ATTENDED, MODEL_TYPE.BACKWARD),
    (PREDICTOR_TYPE.ENVELOPE_ONSET, ATTENTION_TYPE.IGNORED,  MODEL_TYPE.BACKWARD),
]
```

```python
SUBJECTS = helper_functions.fuglsang_get_subjects()

trf_data, _ = helper_functions.load_trfs(
    DATASET_TYPE.FUGLSANG, SUBJECTS, trf_types, trf_dir=FUGLSANG_TRF_SELF_DIR
)
```

```python
# ─────────────────────────────────────────────
# Average all 18 subject TRFs into universal TRFs and save
# ─────────────────────────────────────────────

for trf_type in trf_types:
    model_name = helper_functions.get_trf_model_name(DATASET_TYPE.FUGLSANG, *trf_type)
    print(f'\nProcessing universal TRF for model: {model_name}')

    trfs = trf_data[model_name]
    print("len of trf_list: ", len(trfs))

    universal_trf = AveragedTRF(eelbrain.combine([trf.h_scaled for trf in trfs]).mean('case'))
    print("universal_trf: ", universal_trf)

    new_model_name = helper_functions.get_trf_model_name(DATASET_TYPE.FUGLSANG, *trf_type, generalised=GENERALISATION_TYPE.AVERAGE)
    eelbrain.save.pickle(universal_trf, FUGLSANG_TRF_GENERAL_DIR / f'full_{new_model_name}.pickle')
    print(f'Saved universal TRF for model {model_name} into {new_model_name}')
```

## Hold out

```python
# ─────────────────────────────────────────────
# 1. Train/test split
# ─────────────────────────────────────────────

test_percent     = 0.3
n_test           = int(len(SUBJECTS) * test_percent)
training_subjects = SUBJECTS[:-n_test]
testing_subjects  = SUBJECTS[-n_test:]
print("Training subjects: ", training_subjects)
print("Testing subjects: ", testing_subjects)
```

```python
# ─────────────────────────────────────────────
# Load training TRFs
# ─────────────────────────────────────────────
training_trf_data, _ = helper_functions.load_trfs(
    DATASET_TYPE.FUGLSANG, training_subjects, trf_types, trf_dir=FUGLSANG_TRF_SELF_DIR
)
```

```python
# ─────────────────────────────────────────────
# Average into universal TRFs and save
# ─────────────────────────────────────────────

for trf_type in trf_types:
    model_name = helper_functions.get_trf_model_name(DATASET_TYPE.FUGLSANG, *trf_type)
    print(f'\nProcessing universal TRF for model: {model_name}')

    trfs = training_trf_data[model_name]
    print("len of trf_list: ", len(trfs))

    universal_trf = AveragedTRF(eelbrain.combine([trf.h_scaled for trf in trfs]).mean('case'))
    print("universal_trf: ", universal_trf)

    new_model_name = helper_functions.get_trf_model_name(DATASET_TYPE.FUGLSANG, *trf_type, generalised=GENERALISATION_TYPE.AVERAGE)
    eelbrain.save.pickle(universal_trf, FUGLSANG_TRF_GENERAL_DIR / f'hold_out_{new_model_name}.pickle')
    print(f'Saved universal TRF for model {model_name} into {new_model_name}')
```

## LOOCV - Leave-one-out Cross-validation

```python
# ─────────────────────────────────────────────
# 3. Average into universal TRFs and save (leave-one-out)
# ─────────────────────────────────────────────

all_trf_data, _ = helper_functions.load_trfs(
    DATASET_TYPE.FUGLSANG, SUBJECTS, trf_types, trf_dir=FUGLSANG_TRF_SELF_DIR
)

for trf_type in trf_types:
    model_name     = helper_functions.get_trf_model_name(DATASET_TYPE.FUGLSANG, *trf_type)
    new_model_name = helper_functions.get_trf_model_name(DATASET_TYPE.FUGLSANG, *trf_type, generalised=GENERALISATION_TYPE.AVERAGE)
    print(f'\nProcessing leave-one-out universal TRFs for model: {model_name}')

    all_trfs = all_trf_data[model_name]

    for i, subject in enumerate(SUBJECTS):
        loo_trf_list  = all_trfs[:i] + all_trfs[i+1:]
        loo_averaged_trf = AveragedTRF(eelbrain.combine([trf.h_scaled for trf in loo_trf_list]).mean('case'))

        eelbrain.save.pickle(
            loo_averaged_trf,
            FUGLSANG_TRF_GENERAL_DIR / f'loocv_{subject}_{new_model_name}.pickle'
        )
        print(f'Saved LOO TRF for held-out subject {subject}')
```
