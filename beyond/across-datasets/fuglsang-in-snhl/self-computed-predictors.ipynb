## Create gammatone spectrogram

We start by using a gammatone filterbank which mimics the cochlea to create a 128 band gammatone spectrogram.

```python
def compute_gammatone(save_dir, stimulus_path):
    """Compute and save gammatone spectrogram for a single stimulus."""

    dst = save_dir / f'{stimulus_path.stem}-gammatone.pickle'
    if dst.exists():
        return
    wav = eelbrain.load.wav(stimulus_path)

    # Mix down to mono if stereo
    if 'channel' in [dim.name for dim in wav.dims]:
        wav = wav.mean('channel')

    gt  = eelbrain.gammatone_bank(
        wav,
        GAMMATONE_FREQUENCY_RANGE_LOW,
        GAMMATONE_FREQUENCY_RANGE_HIGH,
        GAMMATONE_SPECTROGRAM_BANDS,
        location='left',
        tstep=0.001
    )
    eelbrain.save.pickle(gt, dst)
    print(f"  ✓ Saved {dst}")


def compute_stimulus_predictors(gt_path, stimulus, env_dir, onset_dir):
    gt     = eelbrain.load.unpickle(gt_path / f'{stimulus}-gammatone.pickle')
    gt_log = (gt + 1).log()
    gt_on  = eelbrain.edge_detector(gt_log, c=30)

    predictors = {
        PREDICTOR_TYPE.ENVELOPE:       (gt_log.sum('frequency'), env_dir),
        PREDICTOR_TYPE.ENVELOPE_ONSET: (gt_on.sum('frequency'),  onset_dir),
    }

    for predictor_type, (predictor, save_dir) in predictors.items():
        p_name = helper_functions.get_predictor_name(predictor_type)
        predictor.name = f'{stimulus}~{p_name}'
        dst = save_dir / f'{stimulus}~{p_name}.pickle'
        if dst.exists():
            continue
        eelbrain.save.pickle(helper_functions.process_predictor(predictor, padded=False), dst)

    print(f"  ✓ Saved predictors for {stimulus}")
```

```python
hi_subjects = helper_functions.get_subjects_snhl(is_hi=True)
nh_subjects = helper_functions.get_subjects_snhl(is_hi=False)

paths = helper_functions.get_stimuli_paths_snhl(SNHL_STIMULI_DIR, hi_subjects)
len(paths)
```

```python
for stim in helper_functions.get_stimuli_paths_snhl(SNHL_STIMULI_DIR, hi_subjects):
    gt_save_dir = SNHL_GAMMATONE_DIR / stim['subject'] / stim['role']
    gt_save_dir.mkdir(parents=True, exist_ok=True)
    compute_gammatone(gt_save_dir, stim['path'])
```

```python
for stim in helper_functions.get_stimuli_paths_snhl(SNHL_STIMULI_DIR, nh_subjects):
    gt_save_dir = SNHL_GAMMATONE_DIR / stim['subject'] / stim['role']
    gt_save_dir.mkdir(parents=True, exist_ok=True)
    compute_gammatone(gt_save_dir, stim['path'])
```

```python
for stim in helper_functions.get_stimuli_paths_snhl(SNHL_STIMULI_DIR, hi_subjects):
    gt_path  = SNHL_GAMMATONE_DIR / stim['subject'] / stim['role']
    env_dir  = SNHL_ENVELOPES_DIR / stim['subject'] / stim['role']
    onset_dir = SNHL_ONSET_DIR    / stim['subject'] / stim['role']
    
    env_dir.mkdir(parents=True, exist_ok=True)
    onset_dir.mkdir(parents=True, exist_ok=True)
    
    compute_stimulus_predictors(gt_path, stim['stem'], env_dir, onset_dir)
```

```python
for stim in helper_functions.get_stimuli_paths_snhl(SNHL_STIMULI_DIR, nh_subjects):
    gt_path  = SNHL_GAMMATONE_DIR / stim['subject'] / stim['role']
    env_dir  = SNHL_ENVELOPES_DIR / stim['subject'] / stim['role']
    onset_dir = SNHL_ONSET_DIR    / stim['subject'] / stim['role']
    
    env_dir.mkdir(parents=True, exist_ok=True)
    onset_dir.mkdir(parents=True, exist_ok=True)
    
    compute_stimulus_predictors(gt_path, stim['stem'], env_dir, onset_dir)
```
